```python
##HANDLING CATEGORICAL COLUMNS##
```

```python
def marital_status(row):
    if row['marital_Divorced'] == 1:
        return 'Divorced'
    elif row['marital_Married'] == 1:
        return 'Married'
    elif row['marital_Single'] == 1:
        return 'Single'
    elif row['marital_Together'] == 1:
        return 'Together'
    elif row['marital_Widow'] == 1:
        return 'Widow'
    else:
        return 'Unknown'
df['Marital'] = df.apply(marital_status, axis=1)
```

```python
##VISUALIZING RELATIONSHIPS##
```

```python
sns.barplot(x='Marital', y='MntTotal', data=df)
sns.set(rc={'figure.figsize':(6,6)})
plt.xlabel('Marital status')
plt.ylabel('MntTotal')
plt.show()
```

```python
def get_relationship(row):
    if row['marital_Married'] ==1:
        return 1
    elif row['marital_Together'] == 1:
        return 1
    else:
        return 0
df['relationship'] = df.apply(get_relationship, axis=1)
df.head()    
```

```python
##APPLYING KMEANS CLUSTERING##
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_for_clustering = ['Income', 'MntTotal', 'relationship']
df_scaled = df.copy()
df_scaled[cols_for_clustering] = scaler.fit_transform(df[cols_for_clustering])
df_scaled[cols_for_clustering].describe()
```

```python
X = df_scaled[cols_for_clustering]
inertia_list = []
for K in range(2,10):
    inertia = KMeans(n_clusters=K, random_state=7).fit(X).inertia_
    inertia_list.append(inertia)

```

```python
plt.figure(figsize=[7,5])
plt.plot(range(2,10), inertia_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_list = []
for K in range(2,10):
    model = KMeans(n_clusters = K, random_state=7)
    clusters = model.fit_predict(X)
    s_avg = silhouette_score(X, clusters)
    silhouette_list.append(s_avg)

plt.figure(figsize=[7,5])
plt.plot(range(2,10), silhouette_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Silhouette Score vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
model = KMeans(n_clusters=4, random_state = 7)
model.fit(df_scaled[cols_for_clustering])
df_scaled['Cluster'] = model.predict(df_scaled[cols_for_clustering])
```

```python
df['Cluster'] = df_scaled.Cluster
df.groupby('Cluster')[cols_for_clustering].mean()
```

```python
mnt_data = df.groupby('Cluster')[cols_mnt].mean().reset_index()
mnt_data.head()
```

```python
##VISUALIZING CLUSTERS##
```

```python
melted_data = pd.melt(mnt_data, id_vars="Cluster", var_name="Product", value_name="Consumption")
plt.figure(figsize=(12, 6))
sns.barplot(x="Cluster", y="Consumption", hue="Product", data=melted_data)
plt.title("Consumption by Cluster")
plt.xlabel("Cluster")
plt.ylabel("Consumption")
plt.xticks(rotation=0)  
plt.legend(title="Product", loc="upper right")

plt.show()
```

```python
cluster_sizes = df.groupby('Cluster')[['MntTotal']].count().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(x='Cluster', y='MntTotal', data=cluster_sizes)
plt.title('Cluster sizes')
plt.xlabel('Cluster')
plt.ylabel('MntTotal')
```
